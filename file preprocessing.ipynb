```python
import os
import json
import pandas as pd
```

```python

# 작업 경로 설정
src = r"한국인 피부상태 측정 데이터/데이터/Training/02.라벨링데이터/3. 스마트폰"

# 데이터를 저장할 리스트
data = []

# 모든 하위 디렉터리 및 파일 탐색
for root, _, files in os.walk(src):
    for file_name in files:
        if file_name.endswith('.json'):  # JSON 파일만 처리
            file_path = os.path.join(root, file_name)
            
            # JSON 파일 읽기
            with open(file_path, 'r', encoding='utf-8') as f:
                json_data = json.load(f)
            
            # 필요한 정보만 추출
            info = json_data.get('info', {})
            images = json_data.get('images', {})
            annotations = json_data.get('annotations', {})
            equipment = json_data.get('equipment', {})

            # 'equipment'가 None일 경우 빈 딕셔너리로 처리
            if equipment is None:
                equipment = {}

            # 추출한 정보를 하나의 딕셔너리로 저장
            data.append({
                'filename': str(file_name).split('.')[0],
                'id': info.get('id'),
                'gender': info.get('gender'),
                'age': info.get('age'),
                'date': info.get('date'),
                'skin_type': info.get('skin_type'),
                'sensitive': info.get('sensitive'),
                'device': images.get('device'),
                'angle': images.get('angle'),
                'facepart': images.get('facepart'),
                'bbox': images.get('bbox'),
                'acne': annotations.get('acne'),
                'pigmentation_count': equipment.get('pigmentation_count', 0)  # 기본값 0
            })

# 리스트를 데이터프레임으로 변환
df = pd.DataFrame(data)

# 결과 출력
print(df.head())
```

```python
import os
import json
import shutil

src = r"한국인 피부상태 측정 데이터/데이터/Training/02.라벨링데이터/3. 스마트폰"

# 데이터를 저장할 리스트
data = []

# 모든 하위 디렉터리 및 파일 탐색
for root, _, files in os.walk(src):
    for file_name in files:

        if file_name.endswith('.json'):  # JSON 파일만 처리
            file_path = os.path.join(root, file_name)  # 전체 경로 생성
        
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)  # JSON 데이터를 Python 객체로 변환
        
         
        # 데이터 수정
        if "info" in data and "skin_type" in data["info"]:
            skin_type = data["info"]["skin_type"]
            if skin_type in [2, 3]:
                data["info"]["skin_type"] = 1
                print('건성으로')
            elif skin_type in [4, 5]:
                data["info"]["skin_type"] = 0
                print('지성으로')
        
        # 수정된 데이터를 JSON 파일로 저장
        with open(file_path, 'w', encoding='utf-8') as file:
            json.dump(data, file, ensure_ascii=False, indent=4)
```
